```python
# eda.ipynb

# 1Ô∏è‚É£ Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option("display.max_columns", None)
sns.set(style="whitegrid")
```

```python
# 2Ô∏è‚É£ Load data
df = pd.read_csv("data/products_clean.csv")
print("‚úÖ Data Loaded")
print(df.shape)
df.head()

```

```python
# 3Ô∏è‚É£ Basic Info
print("\nüìä Dataset Info:")
print(df.info())

print("\nüîπ Missing Values:")
print(df.isnull().sum())

print("\nüìâ Price Stats (numeric only):")
print(df["price_value"].describe())

```

```python
# 4Ô∏è‚É£ Missing value heatmap
plt.figure(figsize=(8,4))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()

```

```python
# 5Ô∏è‚É£ Top Locations
plt.figure(figsize=(8,4))
df["location"].value_counts().head(10).plot(kind="bar", color="skyblue")
plt.title("Top 10 Locations by Product Count")
plt.xlabel("Location")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

```

```python
# 6Ô∏è‚É£ Seller Frequency
plt.figure(figsize=(8,4))
df["seller"].value_counts().head(10).plot(kind="bar", color="orange")
plt.title("Top 10 Sellers")
plt.xlabel("Seller")
plt.ylabel("Products")
plt.xticks(rotation=45)
plt.show()

```

```python
# 7Ô∏è‚É£ Price Distribution (numeric only)
plt.figure(figsize=(8,5))
sns.histplot(df["price_value"].dropna(), bins=20, kde=True)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.show()


```

```python
# 8Ô∏è‚É£ Price Boxplot by Location
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="location", y="price_value")
plt.xticks(rotation=45)
plt.title("Price Distribution by Location")
plt.show()

```

```python
# 9Ô∏è‚É£ Keyword distribution
plt.figure(figsize=(6,4))
df["keyword"].value_counts().plot(kind="bar", color="green")
plt.title("Products by Search Keyword")
plt.ylabel("Count")
plt.show()
```
